```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Select features and target variable
x = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
y = df['Survived']

# Convert 'Sex' column to numerical using Label Encoding
le = LabelEncoder()
x['Sex'] = le.fit_transform(x['Sex'])

# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=0)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
# Reload the test data to ensure 'PassengerId' is present
test_url = 'https://github.com/robitussin/CCMACLRL_EXERCISES_OLD/blob/f5fdfba1aabbbafcf2fe49ca6e1c8f92f69d1458/datasets/titanic/test.csv?raw=true'
dt = pd.read_csv(test_url)

# Store PassengerId before any potential row dropping or index changes
passenger_ids_original = dt['PassengerId']

# Handle missing values in the test set by filling with the mean for Age and mode for Embarked
# Cabin has too many missing values, so we will not use it.
dt['Age'].fillna(dt['Age'].mean(), inplace=True)
dt['Fare'].fillna(dt['Fare'].mean(), inplace=True) # Also fill missing Fare
dt['Embarked'].fillna(dt['Embarked'].mode()[0], inplace=True)


# Select the same features as used for training and create a copy to avoid SettingWithCopyWarning
x_test_processed = dt[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']].copy()

# Apply the same LabelEncoder to the 'Sex' column in the test data
# Assuming 'le' object from training phase is available
x_test_processed['Sex'] = le.transform(x_test_processed['Sex'])

# Ensure the order of columns in x_test_processed matches x_train
x_test_processed = x_test_processed[x_train.columns]


# Check for and handle any remaining missing values
x_test_processed.dropna(inplace=True)

# Select the passenger IDs corresponding to the rows that remain in x_test_processed
passenger_ids_aligned = passenger_ids_original.loc[x_test_processed.index]


y_pred = model.predict(x_test_processed)

# Create a submission DataFrame
submission_df = pd.DataFrame({
    'PassengerID': passenger_ids_aligned,
    'Survived': y_pred
})

# Save the submission DataFrame to a CSV file
submission_df.to_csv('submission_file.csv', index=False)
print("Submission file created: submission_file.csv")
```
